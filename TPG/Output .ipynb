```python
import os
from sys import argv, stdout, stderr
from ROOT import TCanvas, TLegend, gROOT, gStyle, TFile, TIter, gSystem, TTree, TH1F, TPaveStats, TGraph, TH2F
import sys
gROOT.SetStyle("Plain")
gROOT.SetBatch(True)
from array import array
```

```python
h = []
o = []

with open('hlsoutput.txt') as f:
    for line in f.readlines():
        word = line.split()
        if word!=[]:
            h.append(int(word[0]))

with open('output.txt') as f1:
    for line1 in f1.readlines():
        word1 = line1.split()
        if word1!=[]:
            o.append(int(word1[0]))
```

```python
gStyle.SetOptStat(0)
canvas = TCanvas("canvas","canvas",800,800)
legend = TLegend(0.1,0.7,0.48,0.9)
legend.SetHeader("570 inputs","C")
h2 = TH2F("h2","",25000,0,25000,25000,0,25000)
h2.GetXaxis().SetTitle('HLS Output')
h2.GetYaxis().SetTitle('CMSSW Output')
for i in range(len(h)):
    h2.Fill(h[i], o[i])
h2.Draw("COL")
legend.Draw()
canvas.Draw()
canvas.SaveAs('HLSvsCMSSW.png')
```
