```python
import os
from sys import argv, stdout, stderr
from ROOT import TCanvas, TLegend, gROOT, gStyle, TFile, TIter, gSystem, TTree, TH1F, TPaveStats, TGraph, TH2F
import sys
gROOT.SetStyle("Plain")
gROOT.SetBatch(True)
from array import array
```

```python
r = []
l = []
with open('hlsoutput.txt') as f:
    for line in f.readlines():
        word = line.split(':')
        r.append(int(word[0]))
        l.append(int(word[1]))
        e.append(int(word[2]))
```

```python
gStyle.SetOptStat(0)
canvas = TCanvas("canvas","canvas",800,800)
#canvas.SetLogy()
legend = TLegend(0.52,0.7,0.8,0.8)
f1 = TFile('ZSkimTree.root')
t1 = f1.Get('zskimtree/TreeR')
f = TFile('ZSkimTreeNoRec.root')
t = f.Get('zskimtree/TreeR')
z1 = t1.GetBranch("ZMass")
z = t.GetBranch("ZMass")
h1 = TH1F("MET Ratio", "MET Ratio", 30, 0, 300)
h1.SetLineColor(2)
h1.GetXaxis().SetTitle('MET (GeV)')
h1.GetYaxis().SetTitle('Events / 1GeV')
h = TH1F("MET", "MET", 30, 0, 300)
legend.AddEntry(h,"No Recovery","l")
legend.AddEntry(h1,"With Recovery","l")
nentries1 = t1.GetEntries()
nentries = t.GetEntries()
h2 = TH2F("h2","",300,0,300,300,0,300)

met1, met = array( 'f' ), array( 'f' )

evmet = {}
evmet1 = {}

for entry1 in t1:
    if entry1.evt in e:
        h1.Fill(entry1.MET)
        #met1.append(entry1.MET)
        evmet1[entry1.evt] = entry1.MET
#h1.Draw("hist SAME")

for entry in t:
    if entry.evt in e:
        h.Fill(entry.MET)
        #met.append(entry.MET)
        evmet[entry.evt] = entry.MET
        
h1.Divide(h)        
h1.Draw("hist")

#h2.GetXaxis().SetTitle('MET No Recovery(GeV)')
#h2.GetYaxis().SetTitle('MET Recovery(GeV)')
#for i in range(len(met)):
#    h2.Fill(met[i], met1[i])
#h.Draw("hist SAME")
#h1.Add(h, -1)
#h1.Draw("hist")
#legend.Draw()

#h2.Draw("COL")

#g = TGraph(n, met1, met)
#g.Draw("acp") 

canvas.Draw()
canvas.SaveAs('METvsMET.png')
```

```python
canvas = TCanvas("canvas","canvas",800,800)
h2 = TH2F("h2","",300,0,300,300,0,300)
h2.GetXaxis().SetTitle('MET No Recovery(GeV)')
h2.GetYaxis().SetTitle('MET Recovery(GeV)')

for i in e:
    h2.Fill(evmet[i], evmet1[i])
    
h2.Draw("COL")
canvas.Draw()
canvas.SaveAs('METvsMET.png')    
```
